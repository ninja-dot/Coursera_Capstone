## Latitudes and Longitudes

```python
import pandas as pd # library for data analysis
import numpy as np # library to handle data in a vectorized manner

link = "http://cocl.us/Geospatial_data"
df1 = pd.read_csv(link)

df1.head()
```

```python
df1.shape
```

```python
df1.columns = ['PostalCode','Latitude','Longitude']

cols = df1.columns.tolist()
cols
```

```python
link = "df_geolink.csv" #csv created in part-1
df = pd.read_csv(link,index_col=0)
df.head()
```

## Merging df,df1 data frames with Postal code

```python
df_new = pd.merge(df, df1, on='PostalCode')
df_new.head(10)
```

```python
df_new.to_csv(r'df_fin.csv') #saving the file as a csv
```
